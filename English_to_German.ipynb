```python
from deep_translator import GoogleTranslator
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
class DictionaryApp:
    def __init__(self):
        self.history = []

    def translate(self, text, target_lang='de'):
        if not text:
            return None, "Please enter some text."
        try:
            translated = GoogleTranslator(source='en', target=target_lang).translate(text)
            entry = {'original': text, 'translated': translated}
            self.history.append(entry)
            return translated, None
        except Exception as e:
            return None, str(e)
```

```python
def run_widget_mode():
    app = DictionaryApp()
    
    title = widgets.HTML("<h2>ğŸ‡©ğŸ‡ª Interactive English-German Dictionary</h2>")
    
    input_box = widgets.Text(
        placeholder='Type English text here...',
        description='English:',
        layout=widgets.Layout(width='500px')
    )
    
    translate_btn = widgets.Button(
        description='Translate',
        button_style='primary',
        icon='language'
    )
    
    output_area = widgets.Output(layout={'border': '1px solid #ddd', 'padding': '10px'})
    
    def on_click_translate(b):
        text = input_box.value
        with output_area:
            clear_output()
            if not text:
                print("Please enter text first.")
                return
            
            print("Translating...")
            result, error = app.translate(text)
            
            clear_output()
            
            if error:
                print(f"Error: {error}")
            else:
                display(
                    widgets.HTML(f"""
                        <div style="font-size: 16px; line-height: 1.5;">
                            <strong>English:</strong> {text}<br>
                            <strong>German:</strong>
                            <span style="color: #0066cc; font-size: 20px;">{result}</span>
                        </div>
                    """)
                )

    translate_btn.on_click(on_click_translate)
    input_box.on_submit(lambda change: on_click_translate(None))

    ui = widgets.VBox([title, input_box, translate_btn, output_area])
    display(ui)
```

```python
run_widget_mode()
```
